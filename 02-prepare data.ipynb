## Split ingredient name and quantity

Separate ingredient name from quantity (liquids)

```python
# Function to split ingredient name and quantity for liquid ingredients only
def separate_liquid_ingredients(x, type):
    ''' 
    x: data input
    type: regex match group
    - If 1, return quantity
    - If 2, return name
    '''
    # Define match for ingredients
    match = r"(.*OUNCES |.*OUNCE |.*TEASPOONS |.*TEASPOON |.*DASHES |.*DASH |.*DROPS |.*DROP)(.*)"
    if re.search(match, x, re.IGNORECASE) is not None:
        output = re.search(match, x, re.IGNORECASE).group(type)
        return output
```

```python
# Select rows with liquid ingredients only
criteria_1 = df_recipes['recipe_ingredient'].str.contains(
    'ounce|teaspoon|dash|drop', case=False) == True

# Split ingredients and write quantity
df_recipes.loc[criteria_1, 'ingredient_quantity'] = (df_recipes
                                                     .loc[criteria_1, 'recipe_ingredient']
                                                     .apply(lambda x: separate_liquid_ingredients(x, 1)))

# Split ingredients and write name
df_recipes.loc[criteria_1, 'ingredient_name'] = (df_recipes
                                                 .loc[criteria_1, 'recipe_ingredient']
                                                 .apply(lambda x: separate_liquid_ingredients(x, 2)))
```

Separate ingredients name from quantity (non-liquid)

```python
# Function to split ingredient name and quantity for non-liquid ingredients only
def separate_other_ingredients(x, type):
    ''' 
    x: data input
    type: regex match group
    - If 1, return quantity
    - If 2, return name
    '''
    match = r"(GARNISH: \d|GARNISH: |\d )(.*)"
    if re.search(match, x, re.IGNORECASE) is not None:
        output = re.search(match, x, re.IGNORECASE).group(type)
        return output
```

```python
# Select rows with non-liquid ingredients only
criteria_1 = df_recipes['recipe_ingredient'].str.contains(
    'ounce|teaspoon|dash|drop', case=False) == False

# Split ingredients and write quantity
df_recipes.loc[criteria_1, 'ingredient_quantity'] = (df_recipes
                                                     .loc[criteria_1, 'recipe_ingredient']
                                                     .apply(lambda x: separate_other_ingredients(x, 1)))
# Split ingredients and write name
df_recipes.loc[criteria_1, 'ingredient_name'] = (df_recipes
                                                 .loc[criteria_1, 'recipe_ingredient']
                                                 .apply(lambda x: separate_other_ingredients(x, 2)))
```

Our simple search will start with matching the parent category and then proceed to match the children category. This makes sense most of the time ("Tequila" , then --> "Blanco") but sometimes it doesn't ("Bourbon" is often named by itself, not "Whiskey --> Bourbon"). <BR><BR>
    
Therefore, we will need to search some subcategories by themselves.

```python
# Define subcategories to search by themselves
search = ['Bourbon', 'Rye', 'Armagnac', 'Calvados', 'Cognac', 'Eau De Vie', 'Grappa', 'Pisco', 'Scotch']

# Loop through dictionary and add "search" flag to specified subcategories
for category in dict_categories:
    
    # Loop through all key value pairs in a category
    for cat_key, cat_value in list(category.items()):
        if cat_key == 'subcategories':
            for subcat in cat_value:
                for subcat_key, subcat_value in list(subcat.items()):
                    if subcat_value in search:
                        subcat['search'] = True
```

## Simple match for spirits - cat 2 and cat 3

A lot specific spirits have their category identified in the ingredient name itself "Rittenhouse 100 Rye" --> "Rye", so a clever list should be able to capture the category of many spirits

Create function to conduct simple match to our redefined product hierarchy above

```python
# Function to search ingredients for any of the categories above present
def match_bottles_simple(x):

    match_dict = {}
    matched = False
    
    match_dict['ingredient_cat_1'] = np.nan
    match_dict['ingredient_cat_2'] = np.nan
    match_dict['ingredient_cat_3'] = np.nan
    
    # Loop through all categories
    for category in dict_categories:
        # Loop through category dictionary
        for cat_key, cat_value in list(category.items()):
            # Match category name
            if cat_key == 'category_name_alt':
                if re.search(cat_value, x, re.IGNORECASE) is not None:
                    match_dict['ingredient_cat_1'] = 'Spirits/Liqueurs'
                    match_dict['ingredient_cat_2'] = cat_value
                    # Return matched to true, so that we know to now begin searching the subcategories
                    matched = True
            
            # Search the same subcategory of that category
            if cat_key == 'subcategories' and matched == True and len(cat_value) > 0:
                for subcat in cat_value:
                    for subcat_key, subcat_value in list(subcat.items()):
                        if subcat_key == 'subcategory_name_alt':
                            if re.search(subcat_value, x, re.IGNORECASE) is not None:
                                match_dict['ingredient_cat_3'] = subcat_value
                                return match_dict
                            
            
            # No matches yet...so try the reverse order of matching
            if cat_key == 'subcategories' and len(cat_value) > 0:
                for subcat in cat_value:
                    for subcat_key, subcat_value in list(subcat.items()):
                        if subcat_key == 'search' and subcat_value == True:
                            if re.search(list(subcat.values())[0], x, re.IGNORECASE) is not None:
                                match_dict['ingredient_cat_1'] = 'Spirits/Liqueurs'
                                match_dict['ingredient_cat_2'] = list(category.values())[0]
                                match_dict['ingredient_cat_3'] = list(subcat.values())[0]                            
                                return match_dict
            
        matched = False
    return match_dict
        
# Test            
match_bottles_simple('Rittenhouse Rye')
```

Create function to fuzzy match exact ingredient (bottles) names

```python
ingredients = df_ingredients['product_name'].tolist()
subcategory_list = df_ingredients['subcategory_name_alt'].tolist()
category_list = df_ingredients['category_name_alt'].tolist()


def match_bottles(x):
    # Empty dict to capture the full bottle match
    full_match = {}
    # Empty list to store scores from matching
    scores = []

    # loop trough all bottles in database and append set_ratio score (farily loose score)
    for item in ingredients:
        scores.append(fuzz.token_set_ratio(x, item))

    # Threshold for match
    if max(scores) >= 70:

        # check to see how many "max" matches exist
        max_scores = [x[0] for x in enumerate(scores) if x[1] == max(scores)]

        # If we have more than 1 max match, there has been a tie and we need to break it with
        # another round of matching
        if len(max_scores) > 1:

            # Create subset of all tied ingredients and categories
            subset_ingredients = [ingredients[i] for i in max_scores]
            subset_ingredients = [str(i) for i in subset_ingredients]
            subset_subcategory_list = [subcategory_list[i] for i in max_scores]
            subset_category_list = [category_list[i] for i in max_scores]

            # Pick a winner based on more strict match
            subset_match = process.extractOne(x, subset_ingredients)[0]

            # Find the index of the winner
            subset_match_index = subset_ingredients.index(subset_match)

            # Write output to dictionary
            full_match['bottle_ingredient_cat_1'] = 'Spirits/Liqueurs'
            full_match['bottle_ingredient_cat_2'] = subset_category_list[subset_match_index]
            full_match['bottle_ingredient_cat_3'] = subset_subcategory_list[subset_match_index]
            full_match['bottle_match'] = subset_ingredients[subset_match_index]
            return full_match
        
        # No ties, so just output the highest previous match
        else:
            full_match['bottle_ingredient_cat_1'] = 'Spirits/Liqueurs'
            full_match['bottle_ingredient_cat_2'] = category_list[scores.index(max(scores))]
            full_match['bottle_ingredient_cat_3'] = subcategory_list[scores.index(max(scores))]
            full_match['bottle_match'] = ingredients[scores.index(max(scores))]
            full_match['bottle_score'] = max(scores)
            return full_match

# Test function
match_bottles('LILLET BLANC')
```

Apply function to dataframe

```python
# Create filters
criteria_1 = df_recipes['ingredient_cat_1'] == 'No match'
criteria_2 = df_recipes['ingredient_cat_1'] == 'Spirits/Liqueurs'

# Create pivot of data to run on unique bottles only
df_bottle_match = df_recipes.loc[criteria_1 | criteria_2, ['ingredient_name']].drop_duplicates('ingredient_name')

# Apply function
df_bottle_match['bottle_match'] = df_bottle_match['ingredient_name'].apply(lambda x: match_bottles(x))
df_bottle_match.head(10)
```
